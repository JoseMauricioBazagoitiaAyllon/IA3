## Tokenización

```python
pregunta_tokens = []
for sentence in preguntas:
  pregunta_tokens.append(sentence.split(' '))

print(pregunta_tokens[0])

respuesta_tokens = []
for sentence in respuestas:
  respuesta_tokens.append(sentence.split(' '))

print(respuesta_tokens[1])
```

```python
def build_token_dict(token_list):
  token_dict = {
      '<PAD>': 0,
      '<START>': 1,
      '<END>': 2
  }

  # Se lee los tokens
  for tokens in token_list:
    for token in tokens:
      if token not in token_dict:
        token_dict[token] = len(token_dict)
  
  return token_dict
```

```python
# tokens preguntas
pregunta_token_dict = build_token_dict(pregunta_tokens)

# tokens respuesta
respuesta_token_dict = build_token_dict(respuesta_tokens)

# Invertimos el diccionario
respuesta_token_dict_inv = {v:k for k,v in respuesta_token_dict.items()}

# Imprimimos nuestros diccinarios
print(pregunta_token_dict)
print(respuesta_token_dict)
print(respuesta_token_dict_inv)
```

```python
encoder_tokens = [['<START>'] + tokens + ['<END>'] for tokens in pregunta_tokens]
decoder_tokens = [['<START>'] + tokens + ['<END>'] for tokens in respuesta_tokens]
output_tokens = [ tokens + ['<END>'] for tokens in respuesta_tokens]

# Se busca la maxima logitud como parametro para las demas 
selector_max_len = max(map(len, encoder_tokens))
styles_max_len = max(map(len, decoder_tokens))

# Se agrega los valores PAD para contar con una longitud pareja
encoder_tokens = [ tokens + ['<PAD>'] * (selector_max_len - len(tokens)) for tokens in encoder_tokens]
decoder_tokens = [ tokens + ['<PAD>'] * (styles_max_len - len(tokens)) for tokens in decoder_tokens]
output_tokens = [ tokens + ['<PAD>'] * (styles_max_len - len(tokens)) for tokens in output_tokens]
```

```python
print(encoder_tokens[0])
print(decoder_tokens[0])
print(output_tokens[0])
```

```python
encoder_input = [list(map(lambda x: pregunta_token_dict[x], tokens)) for tokens in encoder_tokens]
decoder_input = [list(map(lambda x: respuesta_token_dict[x], tokens)) for tokens in decoder_tokens]
output_decoded = [list(map(lambda x: [respuesta_token_dict[x]], tokens)) for tokens in output_tokens]
```

```python
print(encoder_input[0])
print(decoder_input[0])
print(output_decoded[0])
```

# Modelo

## Construir el modelo

```python
model = get_model(
    token_num=max(len(pregunta_token_dict), len(respuesta_token_dict)),
    embed_dim=32,
    encoder_num=2,
    decoder_num=2,
    head_num=4,
    hidden_dim=128,
    dropout_rate=0.05,
    use_same_embed=False,
    embed_weights=None  # Use different embeddings for different languages
)
model.compile('adam', 'sparse_categorical_crossentropy')
model.summary()
```

## Entrenar el modelo

```python
x = [np.array(encoder_input), np.array(decoder_input)]
y = np.array(output_decoded)

model.fit(x, y, epochs=500, batch_size = 64)
```

```python
model.save("completeModel.h5")
```

## Funcion AnswersGenerator


```python
def AnswersGenerator(sentences):
  sentence_tokens = [tokens + ['<END>', '<PAD>'] for tokens in [sentences.split(' ')]]
  print(sentence_tokens)
  tr_input = [list(map(lambda x: pregunta_token_dict[x], tokens)) for tokens in sentence_tokens][0]
  print(tr_input)
  decoded = decode(
      model,
      tr_input,
      start_token = respuesta_token_dict['<START>'],   
      end_token = respuesta_token_dict['<END>'],   
      pad_token = respuesta_token_dict['<PAD>']   
  )
  resp = format(' '.join(map(lambda x: respuesta_token_dict_inv[x], decoded[1:-1])))
  #aqui imprimimos los resultados
  #print("Pregunta : {} ".format(sentences))
  #print("Respuesta :", resp)
  return resp
```
